## PP Adam EoS – plots only (from v2 experiments)

```python
# Plotly rendering setup
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = "plotly_mimetype"

def show_fig(fig):
    """Show fig: try plotly_mimetype+png, then plotly_mimetype, then notebook; last resort use HTML."""
    for renderer in ("plotly_mimetype+png", "plotly_mimetype", "notebook"):
        try:
            fig.show(renderer=renderer)
            return
        except Exception:
            continue
    from IPython.display import display, HTML
    display(HTML(pio.to_html(fig, include_plotlyjs="cdn")))
```

```python
from pathlib import Path
import json
import pandas as pd

OUTPUT_V2 = Path("../output/eos/adam_PP_v2_full")

def lambda_lim(eta, beta1):
    return (2 + 2 * beta1) / ((1 - beta1) * eta)
```

```python
from plotly.subplots import make_subplots

DEFAULT_COLORS = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]

def plot_fig1a_train_loss(lrs, output_dir, colors=None):
    """Train loss vs step, one trace per η."""
    if colors is None:
        colors = DEFAULT_COLORS
    fig = go.Figure()
    for i, eta in enumerate(lrs):
        try:
            df = pd.read_csv(output_dir / f"fig1a_eta_{eta}" / "metrics.csv")
            fig.add_trace(go.Scatter(x=df["step"], y=df["train_loss"], name=f"η={eta}", line=dict(color=colors[i % len(colors)])))
        except Exception as e:
            print(f"Skip {eta}: {e}")
    fig.update_xaxes(title_text="Step")
    fig.update_yaxes(title_text="Train loss")
    fig.update_layout(height=300, title="Train loss (by learning rate)")
    return fig

def plot_fig1a_lambda_A(lrs, output_dir, colors=None):
    """λ_A vs step (log), with λ_lim hlines."""
    if colors is None:
        colors = DEFAULT_COLORS
    fig = go.Figure()
    for i, eta in enumerate(lrs):
        try:
            df = pd.read_csv(output_dir / f"fig1a_eta_{eta}" / "metrics.csv")
            d = df.dropna(subset=["lambda_A"])
            fig.add_trace(go.Scatter(x=d["step"], y=d["lambda_A"], name=f"η={eta}", line=dict(color=colors[i % len(colors)])))
            lim = lambda_lim(eta, 0.9)
            fig.add_hline(y=lim, line_dash="dash", line_color=colors[i % len(colors)], opacity=0.7)
        except Exception as e:
            print(f"Skip {eta}: {e}")
    fig.update_xaxes(title_text="Step")
    fig.update_yaxes(type="log", title_text="Preconditioned sharpness")
    fig.update_layout(height=300, title="Preconditioned sharpness (by learning rate)")
    return fig

def plot_fig1b_train_loss(betas, output_dir, colors=None):
    """Train loss vs step, one trace per β1."""
    if colors is None:
        colors = DEFAULT_COLORS
    fig = go.Figure()
    for i, b1 in enumerate(betas):
        try:
            df = pd.read_csv(output_dir / f"fig1b_beta1_{b1}" / "metrics.csv")
            fig.add_trace(go.Scatter(x=df["step"], y=df["train_loss"], name=f"β1={b1}", line=dict(color=colors[i % len(colors)])))
        except Exception as e:
            print(f"Skip β1={b1}: {e}")
    fig.update_xaxes(title_text="Step")
    fig.update_yaxes(title_text="Train loss")
    fig.update_layout(height=300, title="Train loss (by β₁)")
    return fig

def plot_fig1b_lambda_A(betas, output_dir, colors=None):
    """λ_A vs step (log), with λ_lim hlines."""
    if colors is None:
        colors = DEFAULT_COLORS
    fig = go.Figure()
    for i, b1 in enumerate(betas):
        try:
            df = pd.read_csv(output_dir / f"fig1b_beta1_{b1}" / "metrics.csv")
            d = df.dropna(subset=["lambda_A"])
            fig.add_trace(go.Scatter(x=d["step"], y=d["lambda_A"], name=f"β1={b1}", line=dict(color=colors[i % len(colors)])))
            lim = lambda_lim(1e-4, b1)
            fig.add_hline(y=lim, line_dash="dash", line_color=colors[i % len(colors)], opacity=0.7)
        except Exception as e:
            print(f"Skip β1={b1}: {e}")
    fig.update_xaxes(title_text="Step")
    fig.update_yaxes(type="log", title_text="Preconditioned sharpness")
    fig.update_layout(height=300, title="Preconditioned sharpness (by β₁)")
    return fig

def plot_fig8_train_loss(lrs, output_dir, colors=None):
    """Train loss for Fig 8 LRs."""
    if colors is None:
        colors = DEFAULT_COLORS
    fig = go.Figure()
    for i, eta in enumerate(lrs):
        try:
            df = pd.read_csv(output_dir / f"fig8_eta_{eta}" / "metrics.csv")
            fig.add_trace(go.Scatter(x=df["step"], y=df["train_loss"], name=f"η={eta}", line=dict(color=colors[i % len(colors)])))
        except Exception as e:
            print(f"Skip {eta}: {e}")
    fig.update_layout(height=300, title="Train loss vs step (multiple learning rates)")
    return fig

def plot_fig8_lambda_A(lrs, output_dir, colors=None):
    """λ_A (log) for Fig 8."""
    if colors is None:
        colors = DEFAULT_COLORS
    fig = go.Figure()
    for i, eta in enumerate(lrs):
        try:
            df = pd.read_csv(output_dir / f"fig8_eta_{eta}" / "metrics.csv")
            d = df.dropna(subset=["lambda_A"])
            fig.add_trace(go.Scatter(x=d["step"], y=d["lambda_A"], name=f"η={eta}", line=dict(color=colors[i % len(colors)])))
            fig.add_hline(y=lambda_lim(eta, 0.9), line_dash="dash", line_color=colors[i % len(colors)], opacity=0.7)
        except Exception as e:
            print(f"Skip {eta}: {e}")
    fig.update_yaxes(type="log")
    fig.update_layout(height=300, title="Preconditioned sharpness λ_A vs step")
    return fig

def plot_fig8_lambda_H(lrs, output_dir, colors=None):
    """λ_H (log) for Fig 8."""
    if colors is None:
        colors = DEFAULT_COLORS
    fig = go.Figure()
    for i, eta in enumerate(lrs):
        try:
            df = pd.read_csv(output_dir / f"fig8_eta_{eta}" / "metrics.csv")
            d = df.dropna(subset=["lambda_H"])
            fig.add_trace(go.Scatter(x=d["step"], y=d["lambda_H"], name=f"η={eta}", line=dict(color=colors[i % len(colors)])))
        except Exception as e:
            print(f"Skip {eta}: {e}")
    fig.update_yaxes(type="log")
    fig.update_layout(height=300, title="Hessian sharpness λ_H vs step")
    return fig


def plot_fig8_combined_sharpness(lrs, output_dir, colors=None):
    """Combined plot: λ_H can increase while λ_A stays bounded near λ_lim (edge)."""
    if colors is None:
        colors = DEFAULT_COLORS

    fig = go.Figure()

    fig.add_trace(
        go.Scatter(
            x=[None],
            y=[None],
            mode="lines",
            name="Unconditioned sharpness",
            line=dict(color="#111", width=2, dash="solid"),
            hoverinfo="skip",
        )
    )
    fig.add_trace(
        go.Scatter(
            x=[None],
            y=[None],
            mode="lines",
            name="Preconditioned sharpness",
            line=dict(color="#111", width=2, dash="dot"),
            hoverinfo="skip",
        )
    )

    for i, eta in enumerate(lrs):
        try:
            df = pd.read_csv(output_dir / f"fig8_eta_{eta}" / "metrics.csv")
            c = colors[i % len(colors)]

            dH = df.dropna(subset=["lambda_H"])
            fig.add_trace(
                go.Scatter(
                    x=dH["step"],
                    y=dH["lambda_H"],
                    name=f"η={eta}",
                    legendgroup=f"eta_{eta}",
                    line=dict(color=c, width=2, dash="solid"),
                )
            )

            dA = df.dropna(subset=["lambda_A"])
            fig.add_trace(
                go.Scatter(
                    x=dA["step"],
                    y=dA["lambda_A"],
                    name=f"η={eta}",
                    legendgroup=f"eta_{eta}",
                    showlegend=False,
                    line=dict(color=c, width=2, dash="dot"),
                )
            )

            fig.add_hline(
                y=lambda_lim(eta, 0.9),
                line_dash="dash",
                line_color=c,
                opacity=0.55,
                annotation_text="λ_lim",
                annotation_position="top left",
            )
        except Exception as e:
            print(f"Skip {eta}: {e}")

    fig.update_yaxes(type="log", title_text="Sharpness")
    fig.update_xaxes(title_text="Step")
    fig.update_layout(
        template="plotly_white",
        height=500,
        title="Unconditioned vs preconditioned sharpness over training",
        legend=dict(yanchor="top", y=0.99, xanchor="left", x=1.02),
        margin=dict(r=160),
    )
    return fig


def plot_single_run_dashboard(run_name, output_dir):
    """Loss + λ_H + λ_A + test acc for a single run."""
    df = pd.read_csv(output_dir / run_name / "metrics.csv")
    with open(output_dir / run_name / "metadata.json") as f:
        meta = json.load(f)
    lr, b1 = meta["lr"], meta["beta1"]
    lim = lambda_lim(lr, b1)
    fig = make_subplots(rows=2, cols=1, specs=[[{"secondary_y": True}], [{}]],
                        subplot_titles=("Loss, λ_H, λ_A", "Test accuracy"), shared_xaxes=True)
    fig.add_trace(go.Scatter(x=df["step"], y=df["train_loss"], name="train loss"), row=1, col=1, secondary_y=False)
    d = df.dropna(subset=["lambda_H"])
    fig.add_trace(go.Scatter(x=d["step"], y=d["lambda_H"], name="λ_H"), row=1, col=1, secondary_y=True)
    dA = df.dropna(subset=["lambda_A"])
    fig.add_trace(go.Scatter(x=dA["step"], y=dA["lambda_A"], name="λ_A"), row=1, col=1, secondary_y=True)
    fig.add_hline(y=lim, line_dash="dash", annotation_text="λ_lim", row=1, col=1, secondary_y=True)
    fig.add_trace(go.Scatter(x=df["step"], y=df["test_acc"], name="test acc"), row=2, col=1)
    fig.update_layout(height=450, title=f"Single run: {run_name} — loss, sharpness, and accuracy")
    fig.update_yaxes(title_text="loss", row=1, col=1, secondary_y=False)
    fig.update_yaxes(title_text="sharpness", type="log", row=1, col=1, secondary_y=True)
    return fig

def plot_edge_ratio(runs, output_dir):
    """edge_ratio(t) = λ_A/λ_lim vs step. Y-axis zooms around 1 so variation is visible."""
    fig = go.Figure()
    all_ratios = []
    for r in runs:
        if not (output_dir / r / "metrics.csv").exists():
            continue
        df = pd.read_csv(output_dir / r / "metrics.csv")
        with open(output_dir / r / "metadata.json") as f:
            meta = json.load(f)
        lim = lambda_lim(meta["lr"], meta["beta1"])
        d = df.dropna(subset=["lambda_A"]).copy()
        d["edge_ratio"] = d["lambda_A"] / lim
        all_ratios.extend(d["edge_ratio"].tolist())
        fig.add_trace(go.Scatter(x=d["step"], y=d["edge_ratio"], name=r, mode="lines+markers"))
    fig.add_hline(y=1.0, line_dash="dash", annotation_text="λ_lim")
    if all_ratios:
        y_min, y_max = min(all_ratios), max(all_ratios)
        span = max(y_max - y_min, 0.05)
        mid = (y_min + y_max) / 2
        fig.update_yaxes(range=[mid - span / 2 - 0.02, mid + span / 2 + 0.02])
    fig.update_layout(title="Edge ratio λ_A/λ_lim vs step", xaxis_title="step", yaxis_title="edge ratio", height=400)
    return fig
```

```python
lrs_fig1a = [3e-5, 1e-4, 3.2e-4, 1e-3]
betas_fig1b = [0.64, 0.814, 0.903, 0.95]
lrs_fig8 = [5e-5, 1e-4, 2e-4, 4e-4]
```

### Train loss and sharpness (varying learning rate η)

```python
show_fig(plot_fig1a_train_loss(lrs_fig1a, OUTPUT_V2))
show_fig(plot_fig1a_lambda_A(lrs_fig1a, OUTPUT_V2))
```

### Train loss and sharpness (varying β₁)

```python
show_fig(plot_fig1b_train_loss(betas_fig1b, OUTPUT_V2))
show_fig(plot_fig1b_lambda_A(betas_fig1b, OUTPUT_V2))
```

### Train loss, λ_A, and λ_H (multiple learning rates)

```python
show_fig(plot_fig8_train_loss(lrs_fig8, OUTPUT_V2))
show_fig(plot_fig8_lambda_A(lrs_fig8, OUTPUT_V2))
show_fig(plot_fig8_lambda_H(lrs_fig8, OUTPUT_V2))
show_fig(plot_fig8_combined_sharpness(lrs_fig8, OUTPUT_V2))
```

### Single run: loss, sharpness, and accuracy

```python
show_fig(plot_single_run_dashboard("sanity", OUTPUT_V2))
```

### Edge ratio λ_A/λ_lim vs step

```python
runs = ["sanity"] + [f"fig1a_eta_{e}" for e in lrs_fig1a] + [f"fig1b_beta1_{b}" for b in betas_fig1b] + [f"fig8_eta_{e}" for e in lrs_fig8]
show_fig(plot_edge_ratio(runs, OUTPUT_V2))
```
